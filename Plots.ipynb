```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
filepath = os.path.abspath("log.txt")
lines = open(filepath).readlines()[9:]
N_generals = []
N_traitors = []
Ret =[]
Rand = []
Rev = []
counter = 0
for line in lines[2:]:
    
    line = line.split(" ")
    data = []
    for element in line:
        if element != '':
            data.append(element)
    try:
        counter +=1
        N_generals.append(int(data[0]))
        N_traitors.append(int(data[1]))
        Ret.append(int(data[2])/int(data[4]))
        Rand.append(int(data[5])/int(data[7]))
        Rev.append(int(data[8])/int(data[10]))
    except:
        print(counter)
        print(line)
        break
```

```python
plt.figure(figsize=(12, 6))
plt.plot(N_generals,Ret,label="Retreat-only traitors")
plt.plot(N_generals,Rand,label="Random message traitors")
plt.plot(N_generals,Rev,label="Reverse message traitors")
plt.xlabel("Number of Generals")
plt.ylabel("Rate")
title = "Success rate with 5 Traitors"
plt.title(title)
plt.legend()
plt.savefig(title)
plt.show()
```
